```python
import numpy as np
import matplotlib.pyplot as plt
f"beta = {1/(0.5*2.269):.3f}"
```

```python
snapshot_times = np.array([10,100,1000,10000])
snapshots = np.loadtxt("out/snapshots.txt").reshape(4,32,32)
```

```python
for i,(time,snapshot) in enumerate(zip(snapshot_times, snapshots)):
    plt.subplot(2,2,i+1)
    plt.title(rf"$t = {time}$")
    plt.imshow(snapshot)
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
avg_M,avg_acceptance = np.loadtxt("out/avg_timeseries.txt").T
time = np.arange(10000)
```

```python
plt.figure(figsize=(8,3))

plt.subplot(1,2,1)
plt.plot(time,avg_M)
plt.xlabel("#sweep")
plt.ylabel(r"$\overline{|M|}$")
plt.xlim(0,10000)

plt.subplot(1,2,2)
plt.plot(time,avg_acceptance)
plt.xlabel("#sweep")
plt.ylabel(r"acc. rate")
plt.xlim(0,1500)
plt.show()
```
