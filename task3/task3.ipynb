### 3. Finite-size corrections for the 1D Ising model
The free energy per spin, $\beta f_N \equiv \frac{1}{N} \ln(Z)$, of the one-dimensional Ising model of a chain with $N$ spins reads in the thermodynamic limit $N \to
\infty$ independent of the boundary conditions: 
$$
\beta f_\infty \equiv \lim_{N \to \infty} \frac{1}{N} \ln(Z) = \ln 2 + \ln \cosh(β)\,.
$$

**a)** Calculate the finite-size correction $\beta f_N − \beta f_\infty$ for i) free and ii) periodic boundary conditions.

`see page before`

**b)** Plot both corrections as a function of $N$ for $\beta = 1$ and $\beta = 10$ in the
range $[1, 10]$. For a comparison it is useful to plot $|\beta f_N − \beta f_\infty |$
(why?).

`because f_fbc is strictly negative, while f_pbc is strictly positive. This makes a direct comparison harder.`


```python
import numpy as np
import matplotlib.pyplot as plt

def f_fbc(N,beta):
    return np.abs(- (1/N) * np.log(np.cosh(beta)))
def f_pbc(N,beta):
    return np.abs(1/N * np.log(1+np.tanh(beta)**N))
```

```python
N = np.linspace(1,10,100)
beta = 1
plt.figure(figsize=(9, 4))
plt.subplot(1,2,1)
plt.plot(N,f_fbc(N,beta), label=r"$fbc, \beta = 1$")
plt.plot(N,f_pbc(N,beta), label=r"$pbc, \beta = 1$")
plt.legend()

beta = 10
plt.subplot(1,2,2)
plt.plot(N,f_fbc(N,beta), label=r"$fbc, \beta = 10$")
plt.plot(N,f_pbc(N,beta), label=r"$pbc, \beta = 10$")
plt.legend()
plt.show()
```

**c)** Repeat **b)** with a logarithmic y-scale in the range $[1, 100]$. Why
does the curve for $\beta = 1$ and periodic boundary conditions appear
as a straight line in the limit of large $N$ ? What does that mean
mathematically? How can one understand the analogous curve
for $\beta = 10$?

`the pbc error goes to 0 exponentially in the case of β = 1 because the argument of the ln is exponentially approaching 1. This behaviour still occurs for β = 10, but cannot be seen in the plot because the tanh-term is extremely close to 1 in that case.`

```python
N = np.linspace(1,100,100)
beta = 1

plt.figure(figsize=(9, 4))
plt.subplot(1,2,1)
plt.yscale("log")
plt.plot(N,f_fbc(N,beta), label=r"$fbc, \beta = 1$")
plt.plot(N,f_pbc(N,beta), label=r"$pbc, \beta = 1$")
plt.legend()

beta = 10
plt.subplot(1,2,2)
plt.yscale("log")
plt.plot(N,f_fbc(N,beta), label=r"$fbc, \beta = 10$")
plt.plot(N,f_pbc(N,beta), label=r"$pbc, \beta = 10$")
plt.legend()
plt.show()
```
